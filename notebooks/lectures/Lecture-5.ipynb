### **Connected components**

The following code finds the connected components of an undirected graph. 


```
# we assume that the input is a list of lists A 
# A[i] is a list of neighbors of node i

# 
def cc(A):
  n = len(A)      # number of nodes
  cId = [-1]*n    # everyone's component ID is initialized to -1
  toExplore = []  # list of nodes still to explore from  (this is a queue)
  

  nc = -1         # number of components found
  
  for j in range(n):

    if len(toExplore) == 0 and cID[j] == -1:
      nc = nc + 1            # new component found
      cID[j] = nc
      toExplore.append(j)    #add j in list to explore

    if len(toExplore)>0:
      u = toExplore.pop()    # pop from the top of the list
      n_u = A[u]             # the list of neighbors of u
      for k in len(n_u):
        v = n_u[k]
        if cID[v] == -1:        # neighbor v has not be discovered before
          cID[v] = nc           # mark v with its component ID
          toExplore.append[v]   # add v in the list of nodes to Explore

  return cID
```

### **An in-place implemetation of partitition**

In the lecture we discussed a function that 'partitions' an array around a pivot element. That implementation was using an extra array. Here I give an in-place implementation, taken from [here](https://www.geeksforgeeks.org/python-program-for-quicksort/)


```
# This function takes last element as pivot, places 
# the pivot element at its correct position in sorted 
# array, and places all smaller (smaller than pivot) 
# to left of pivot and all greater elements to right 
# of pivot 

def partition(arr,low,high): 
    i = ( low-1 )         # index of smaller element 
    pivot = arr[high]     # pivot 
  
    for j in range(low , high): 
  
        # If current element is smaller than or 
        # equal to pivot 
        if   arr[j] <= pivot: 
          
            # increment index of smaller element 
            i = i+1 
            arr[i],arr[j] = arr[j],arr[i] 
  
    arr[i+1],arr[high] = arr[high],arr[i+1] 
    return ( i+1 ) 
```

### **Quick-Sort without randomness**

The following cell gives an implementation without randomness. It just uses the above *partition* function, which by default pivots around the last element of the array. 

**Question:**  How would you change the implementation in order to have a random pivot?

```
# The main function that implements QuickSort 
# arr[] --> Array to be sorted, 
# low  --> Starting index, 
# high  --> Ending index 
  
# Function to do Quick sort 
def quickSort(arr,low,high): 
    if low < high: 
  
        # pi is partitioning index, arr[p] is now 
        # at right place 
        pi = partition(arr,low,high) 
  
        # Separately sort elements before 
        # partition and after partition 
        quickSort(arr, low, pi-1) 
        quickSort(arr, pi+1, high) 
```

### **order-k element**

Using the same partition function as above, we now implement a function that finds the k-th order element in a list. Again here we implement a version of the algorithm that does not use randomness. 

```
# the input k below can be in range(0,len(arr))

def kth(arr,low,high, k):

  if low<high:

    pi = partition(arr,low,high)

    if k<pi:
      return kth(arr,low,pi, k)
    else:
      return kth(arr,pi,high, k-pi )

  else:
    return arr[low]   # k must be 0


```
