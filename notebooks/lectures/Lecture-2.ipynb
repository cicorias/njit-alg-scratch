### **Some things about matrices** ###

```
# we will be using numpy to work with two dimensional matrices

import numpy as np    # the 'as np' part gives numpy a shorter name so we can access its methods

# nxn matrix multiplication
# the following function assumes the input matrices have the same square shape

def matMult(A,B):
  n = A.spape[0]
  C = np.zeros([n,n])    # initialize C with zeros
  for i in range(n):
    for j in range(n):
      for k in range(n):
        C[i,j] = C[i,j]+A[i,k]*B[k,j]
  return C

```
