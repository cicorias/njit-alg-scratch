### **Depth First Search**

This is a simple recursive implementation of the depth-first search algorithm. For your mini-project you many need to implement an iterative version.


```
# This is the adjacency list of the graph in figure 3.9 of the book chapter I provided. 
# Letters are replaced by corresponding numbers 

G = [[1],[2,3,4],[3],[],[1,5,6],[2,7], [7,9],[10],[6], [8],  [11],[9]]
```

```
# the following function initializes global arrays
def reset_arrays():

  global visited, clock, start , finish 


  n = len(G)
  visited = [0]*n
  start = [None]*n
  finish = [None]*n
  clock = 0;


# the following function explores the graph out of u
# marks start and finish times 
# and records the component id for each node
def explore(G,u):  

  global visited, clock, cc, start, finish 
  
  #pre-visit
  start[u] = clock
  clock = clock+1

  visited[u] = 1
  for v in G[u]:     # Go over neighbors of u
    if visited[v]==0 :
      explore(G,v)

  #post-visit
  finish[u] = clock
  clock = clock+1

# this does a full dfs
# 
def dfs(G):


  for u in range(n):

    if visited[u]==0:
      explore(G,u)



```

```
# try some values here

reset_arrays()
explore(G,10)
print(visited)
print(start)
print(finish)
```
