## **Question 6**

A brute force algorithm:

```
def printRamanujan(n):
  for x in range(2,n-1):
    y = n -x
    for a in range(1,x):
      b = x -a
      for c in range (1,y):
        d = y -c
        if a**3 + b**3 == c**3 + d**3:
          print ("a = %d, b = %d, c = %d, d = %d" % (a, b, c, d))

printRamanujan(32)
```

Running time: $O(n^3)$
